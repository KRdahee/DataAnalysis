# 의사결정 트리(Decision Tree)

* 의사결정트리(Decision Tree)는 데이터 분석과 패턴 인식에서 널리 사용되는 지도 학습 알고리즘
  - 계층적 구조로서 각각의 노드는 특정 조건에 대한 결정 규칙을 나타냄
  - 가지(branch)로 표현되는 조건에 따라 분기
  - 직관적으로 이해하기 쉽고 시각화하기 용이함
  - 범주형 데이터와 수치형 데이터를 모두 처리 가능
  - 이상치(Outlier)에 상대적으로 영향을 받지 않음
  - 데이터의 스케일링이나 정규화가 필요하지 않음
  - 입력 변수의 중요도를 평가할 수 있으므로, 특성 선택(Feature Selection)을 수행하는 데에도 유용
  - 중요한 특성들을 파악하여 모델의 설명력을 높일 수 있음

## 분류 - `DecisionTreeClassifier()`

* `DecisionTreeClassifier`는 분류를 위한 결정트리 모델
* 두개의 배열 X, y를 입력 받음
  * X는 [n_samples, n_features] 크기의 데이터 특성 배열
  * y는 [n_samples] 크기의 정답 배열
