## 일변량 시계열 데이터

### ARIMA 모델

*   ARIMA는 자기회귀 이동평균 모델로, 시계열 데이터의 특징을 파악하여 예측하는데 사용
*   ARIMA 모델은 자기회귀(AR) 항, 차분(I) 항, 이동평균(MA) 항을 포함
*   AR 항은 이전 시점 값들과 현재 시점의 관계를 나타내는 항으로, 시계열 데이터의 자기상관성을 모델링
*   차분 항은 시계열 데이터의 정상성(stationarity)을 확보하기 위해 이전 시점의 값과 현재 시점의 차이를 계산
*   MA 항은 이동평균 모델로, 이전 시점의 오차(residuals)와 현재 시점의 예측값과의 관계를 모델링
*   ARIMA 모델은 파라미터인 p, d, q를 통해 자기회귀, 차분, 이동평균의 차수 조정 가능
*   ARIMA는 예측력이 뛰어나지만, 계절성을 갖는 시계열 데이터에 적용하기에는 한계가 있음

### ETS 모델

*  ETS(Exponential Smoothing State Space) 모델은 시계열 데이터를 예측하기 위한 모델 중 하나
*  ETS 모델은 시계열 데이터의 추세와 계절성을 고려하여 예측하는데 사용
* 이 모델은 시계열 데이터를 세 가지 요소로 분해
  - Level(수준)
    - 시계열 데이터의 평균 수준을 나타냄
    - 예를 들어, 주식 가격의 평균 수준은 주식의 전반적인 경향을 나타냄
  - Trend(추세)
    - 시계열 데이터의 증가 또는 감소하는 경향을 나타냄
    - 추세는 시계열 데이터의 장기적인 변화를 나타냄
  - Seasonality(계절성)
    - 시계열 데이터의 반복적인 패턴을 나타냄
    - 계절성은 데이터에 주기적인 변동이 있는 경우 관찰
    - 예를 들어, 매년 같은 월에 판매량이 증가하는 경우 계절성 패턴이 있을 수 있음

*  ETS 모델은 이러한 세 가지 요소를 고려하여 다음 시점의 값을 예측
*  모델은 지수평활법(Exponential Smoothing)을 사용하여 데이터의 과거 값들에 가중치를 부여하고, 최근 값에 더 큰 가중치를 부여하여 예측을 수행

### BATS 모델

*  BATS(Bayesian Structural Time Series) 모델은 시계열 데이터의 추세, 계절성 및 잡음 요소를 모델링하여 예측을 수행

*  BATS 모델은 일반적으로 세 가지 요소로 구성

  - Trend(추세)
    - 추세는 시계열 데이터의 장기적인 변화를 나타냄
    - BATS 모델은 추세를 유연한 방식으로 모델링하기 위해 로버스트 추세 컴포넌트를 사용
    - 추세는 시간에 따라 변화할 수 있으며, 추세의 변화는 상태 공간 모델로 표현

  - Seasonality(계절성)
    - 계절성은 시계열 데이터의 주기적인 패턴을 나타냄
    - BATS 모델은 계절성을 장, 중, 단기 성분으로 분해
    - 장기 계절성은 긴 주기의 패턴을 나타내며, 중기와 단기 계절성은 보다 짧은 주기의 패턴을 나타냄
    - 각 계절성 요소는 상태 공간 모델로 모델링

  - 잡음(Noise)
    - 잡음은 모델로 예측할 수 없는 불규칙한 변동을 의미
    - BATS 모델은 잡음을 정규 분포 혹은 스튜던트 t 분포로 모델링
    - 잡음은 예측의 불확실성을 나타내며, 예측 구간의 폭을 결정하는 데 사용

*  베이지안 추론을 기반으로 하며, MCMC(Markov Chain Monte Carlo) 알고리즘을 사용하여 모델 파라미터 및 상태 공간을 추정
*  예측 구간과 예측의 불확실성을 포함한 종합적인 예측 결과를 얻을 수 있음

### Time Series Forest & Basic motions


#### Time Series Forest
  *   시계열 데이터를 분석하기 위한 머신 러닝 알고리즘 중 하나
  *   랜덤 포레스트 알고리즘의 변형으로, 시계열 데이터의 특성을 분석하고 예측하는 데 사용
  *   각 시계열의 여러 구간에서 기초적인 통계치들을 계산한 후, 이를 랜덤 포레스트의 입력 데이터로 사용


#### Basic Motions 데이터셋
  *  스마트 워치를 착용한 네 명의 학생들이 네 가지 활동을 수행하는 과정에서 생성
  *  활동은 걷기, 휴식, 달리기, 배드민턴으로, 스마트 워치는 3D 가속도계와 3D 자이로스코프를 수집
  *  각 참가자는 총 다섯 번의 움직임을 기록하였으며, 데이터는 10초 동안 0.1초마다 샘플링

### RandomIntervalClassifier & PLAID 데이터셋


#### RandomIntervalClassifier

*  Random Interval Spectral Ensemble (RISE)에 기반을 둔 이 모델은 랜덤하게 선택된 시계열의 구간에 대해 특징을 추출하고, 그 특징들로부터 분류 모델을 학습
*  RISE는 일반적으로 복잡한 패턴이나 노이즈가 있는 시계열 데이터에 대해서도 효과적
   

#### PLAID 데이터셋

#### SVC

*  SVM은 이진 또는 다중 클래스 분류 문제를 해결하기 위한 강력한 머신 러닝 알고리즘
*  SVM의 기본 아이디어는 데이터를 고차원 공간에 매핑한 다음, 서로 다른 클래스 사이의 가장 넓은 '거리'를 가지는 초평면(결정 경계)을 찾는 것
*  이러한 결정 경계를 '마진'이라고 부르며, SVM은 마진을 최대화함으로써 모델의 일반화 성능을 높임
*  SVM은 커널 트릭을 이용하여 비선형 분류 문제도 풀 수 있음 → 선형적으로 분리되지 않는 데이터도 고차원 공간에서 선형적으로 분리 가능
    

#### Italy Power Demand 데이터셋

   *  이탈리아의 전력 수요를 시간별로 기록한 시계열 데이터
   *  각 시계열은 24시간 동안의 전력 수요를 나타냄
   *  특정 날짜의 전력 수요 패턴이 주중인지 주말인지를 분류하는 것이 일반적인 예시
   *  이런 분류 작업은 전력 그리드의 운영 및 관리, 에너지 수요 예측 등의 응용 분야에 중요한 역할을 함

### Basic Motions 데이터셋

- 4명의 학생이 스마트 워치를 착용한 채 4가지 활동을 수행
- 스마트 워치는  3D 가속도계와 3D 자이로스코프를 수집
- 데이터 순서는 가속도계 x, y, z, 자이로스코프 x, y, z이며, 걷기, 휴식, 달리기, 배드민턴으로 분류
- 참가자는 총 5회 동작을 기록해야 하며, 데이터는 10초에 한 번씩 10초 동안 샘플링

https://www.timeseriesclassification.com/description.php?Dataset=BasicMotions

### Japanese Vowels 데이터셋

- 9명의 일본 남성 화자가 모음 'a'와 'e'를 말하는 것을 녹음
- 원시 녹음에 '12도 선형 예측 분석'을 적용하여 원래 길이가 7에서 29 사이인 12차원의 시계열
- 분류 작업은 화자를 예측하는 것
- 각 인스턴스는 단일 클래스 레이블 [1...9]가 부착된 12*29 값으로 변환된 발화
- 주어진 훈련 세트는 각 화자당 30개의 발화로 구성되지만, 테스트 세트는 타이밍과 실험 가용성이라는 외부 요인에 따라 화자당 24개에서 88개까지 다양한 분포를 가짐

http://timeseriesclassification.com/description.php?Dataset=JapaneseVowels

### PLAID 데이터셋

- 플러그 부하 기기 식별 데이터 세트로 부하 식별 연구를 위한 것
- 미국 펜실베이니아주 피츠버그의 56개 이상의 가정에 있는 11가지 가전제품 유형에서 30kHz로 샘플링한 전류 및 전압 측정값이 포함 (2013년 여름)
- 각 가전 제품 유형은 제조사/모델이 다양한 수십 개의 서로 다른 사례로 표현
- 각 어플라이언스에 대해 각 상태 전환에 대해 3~6개의 측정값을 수집
- 측정값을 후처리하여 정상 상태 작동과 시동 과도 상태(가능한 경우)를 모두 포함하는 몇 초 길이의 윈도우를 추출
- 에어컨, 소형 형광등, 냉장고, 헤어 드라이어, 노트북, 전자레인지, 세탁기, 전구, 진공 청소기, 선풍기, 히터 등 11가지 가전 제품 유형에 해당
- 각 시계열의 끝에 가장 긴 시계열의 길이만큼 NaN을 추가

http://timeseriesclassification.com/description.php?Dataset=PLAID
