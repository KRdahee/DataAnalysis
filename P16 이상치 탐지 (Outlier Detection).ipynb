# 이상치 탐지 (Outlier Detection)

### 사용할 데이터셋


* Credit Card Fraud Detection
 * Credit Card Fraud Detection 데이터셋은 신용카드 거래 데이터를 기반으로 신용카드 사기를 탐지하는데 사용
 * 정상 거래와 사기 거래의 불균형한 분포를 가지고 있으며, 이상치 탐지 알고리즘은 이 불균형한 분포에서 사기 거래를 식별하는데 활용

* Numenta Anomaly Benchmark (NAB)
 * NAB 데이터셋은 다양한 도메인의 시계열 데이터에서 이상치를 탐지하는 벤치마크 데이터셋
 * 각 데이터셋은 특정 도메인에서 발생하는 이상치 패턴을 모사하고 있으며, 이상치 탐지 알고리즘의 성능을 평가하는데 사용

* Walmart Cleaned Data
 * Walmart Cleaned Data는 월마트의 판매 데이터에서 이상치를 탐지하는 데이터셋
 * 주간 판매량, 가격, 실업률 등의 특성을 가지고 있으며, 이상치 탐지 알고리즘은 판매량이나 가격의 이상치를 탐지하는데 활용

* KDD Cup 1999 Data

 * KDD Cup 1999 Data는 네트워크 트래픽에서 이상치를 탐지하는 데이터셋
 * 다양한 네트워크 연결 및 통신 프로토콜을 포함하고 있으며, 이상한 행동을 나타내는 이상치를 탐지하는데 사용

* Forest Cover Type Prediction

 * Forest Cover Type Prediction은 숲의 토양 조건에 따라 토양 유형을 예측하는 데이터셋
 * 숲의 지리적 및 환경적 특성을 포함하고 있으며, 이상치 탐지 알고리즘은 토양 조건에서 벗어난 이상치를 탐지하는데 활용

## PyOD 라이브러리

파이썬 기반의 오픈 소스 라이브러리로, 이상치 탐지와 이상치 분석을 위한 다양한 기능을 제공

* 다양한 이상치 탐지 알고리즘
  - 대표적으로 k-NN, LOF(Local Outlier Factor), ABOD(Angle-Based Outlier Detection) 등이 있음
  - 각각의 알고리즘은 다양한 데이터 유형과 패턴을 탐지하는 데 사용
* 간편한 사용법
  - 사용하기 쉽고 직관적인 API를 제공하여 신속하게 이상치 탐지 모델 구축 가능
  - 데이터 과학자나 머신러닝 엔지니어가 효율적으로 이상치 탐지 모델을 개발하고 평가할 수 있도록 도와줌
* 다양한 데이터 유형과 도메인에 대한 지원
  - 다양한 유형의 데이터에 대해 이상치 탐지 수행 가능
  - 수치형 데이터, 텍스트 데이터, 이미지 데이터 등 다양한 도메인에서 이상치 탐지 가능
* 확장성과 유연성
  - 다른 파이썬 머신러닝 라이브러리와의 통합이 용이하며, 다른 라이브러리와 함께 사용하여 더 복잡한 이상치 탐지 작업 수행 가능
  - 사용자 정의 가능한 매개변수를 통해 모델의 동작을 조정할 수 있어 유연한 탐지 방법 구현 가능
* 활용 분야의 다양성
  - 데이터의 품질 향상, 보안 강화, 사기 탐지, 이상 거래 감지 등 다양한 분야에서 활용 가능
  - 이상치 탐지를 통해 비정상적인 패턴이나 동작을 식별하고, 신속하게 대응하고 예방할 수 있는 잠재적인 가치를 제공

## LOF (Local Outlier Factor)

LOF는 주변 이웃들과 비교하여 각 데이터 포인트의 이상치 점수를 계산하는 비지도 학습 기반 이상치 탐지 알고리즘

* 장점:
  * 국지적인 밀도 기반 이상치 탐지 방법으로, 데이터 분포의 밀도 변화에 더 잘 대응할 수 있음
  * 이상치 판단에 있어서 비교적 간단하고 직관적인 개념
* 단점:
  * 데이터셋이 매우 큰 경우 계산 비용이 높을 수 있음
  * 이상치가 아닌 데이터 포인트들 사이의 밀도가 서로 다를 때, 적절한 결과를 제공하지 못할 수 있음

## Comparing All Implemented Outlier Detection Models

* Linear Models for Outlier Detection
  - PCA(Principal Component Analysis)
  - MCD(Minimum Covariance Determinant)
  - OCSVM(One-Class Support Vector Machines)
* Proximity-Based Outlier Detection Models
  - LOF(Local Outlier Factor)
  - CBLOF(Clustering-Based Local Outlier Factor)
  - kNN(k Nearest Neighbors)
  - Median kNN Outlier Detection
  - HBOS(Histogram-based Outlier Score)
* Probabilistic Models for Outlier Detection
  - ABOD(Angle-Based Outlier Detection)
* Outlier Ensembles and Combination Frameworks
  - Isolation Forest
  - Feature Bagging
  - LSCP
