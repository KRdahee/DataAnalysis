# 인공신경망(Artificial Neural Network)

- MLP는 다층 퍼셉트론(Multilayer Perceptron)의 약어로, 인공신경망(Artificial Neural Network)의 한 종류
- 다층 퍼셉트론은 입력층(input layer), 은닉층(hidden layer), 출력층(output layer)으로 구성되며, 각 층은 여러 개의 뉴런(neuron)으로 이루어짐

MLP의 특징
- 유연성: 다층 퍼셉트론은 다양한 유형의 문제에 적용할 수 있으며, 입력 데이터의 복잡한 패턴 학습 가능
- 비선형성: 은닉층의 활성화 함수로 비선형 함수를 사용함으로써, 비선형 문제 해결 가능
- 전방향 신경망: 입력에서 출력 방향으로 정보가 전달되는 단방향 구조를 가지고 있음
- 지도학습: 주로 지도학습 문제에 적용되며, 입력과 출력 데이터 사이의 관계를 학습
- 오차 역전파 알고리즘: MLP는 역전파(backpropagation) 알고리즘을 사용하여 가중치를 조정하고 오차를 최소화

MLP의 장점
- 다양한 문제에 적용 가능: MLP는 다양한 분야에서 사용되며, 이미지 및 음성 인식, 자연어 처리 등 다양한 문제에 적용할 수 있음
- 병렬 처리: MLP는 병렬 처리에 적합하며, 고성능 컴퓨터나 그래픽 처리 장치(GPU)를 이용하여 빠른 연산이 가능

MLP의 단점
- 과적합 문제: MLP는 층의 수와 뉴런의 수가 많아질수록 과적합(overfitting) 문제가 발생할 수 있으므로 적절한 정규화 기법을 적용하여 이를 완화해야 함
- 데이터 전처리 필요: MLP는 입력 데이터의 전처리에 민감할 수 있으며, 데이터의 정규화나 특징 선택 등의 과정이 필요할 수 있음
- 훈련 시간과 계산 비용: MLP는 대규모의 데이터셋이나 복잡한 네트워크 구조에서 훈련에 많은 시간과 계산 비용이 소요될 수 있음
