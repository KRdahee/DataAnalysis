# 추천 시스템 (Recommender System)

*   추천 시스템은 크게 두가지로 구분
  *   컨텐츠 기반 필터링(content-based filtering)
  *   협업 필터링(collaborative filtering)
*   두가지를 조합한 hybrid 방식으로도 가능
*   컨텐츠 기반 필터링은 지금까지 사용자의 이전 행동과 명시적 피드백을 통해 사용자가 좋아하는 것과 유사한 항목을 추천
*   협업 필터링은 사용자와 항목간의 유사성을 동시에 사용해 추천

## Surprise

- Surprise 는 명시적 평가 데이터를 다루는 추천 시스템을 구축하고 분석하기 위한 Python scikit
- 다양한 모델과 데이터를 제공하고, - Scikit-learn과 사용방법이 유사함
- 사용자는 알고리즘의 모든 세부 사항을 수정 가능
- 사용자는 기본 제공 데이터 세트(Movielens, Jester)와 자체 사용자 지정 데이터 세트를 모두 사용 가능
- 기준선 알고리즘 , 이웃 방법 , 행렬 분해 기반( SVD , PMF , SVD++ , NMF ) 등 다양한 즉시 사용 가능한 예측 알고리즘을 제공
- 다양한 유사도 측정 (cosine, MSD, pearson...)이 내장
- 새로운 알고리즘 아이디어를 쉽게 구현 가능
- 알고리즘의 성능을 평가, 분석 및 비교하는 도구를 제공

## 컨텐츠 기반 필터링 (Content-based Filtering)

*   컨텐츠 기반 필터링은 이전의 행동과 명시적 피드백을 통해 좋아하는 것과 유사한 항목을 추천
* Netflix, YOUTUBE 등과 같은 OTT에서 내가 시청한 컨텐츠 목록과 다른 사용자의 목록을 비교해 비슷한 취향의 사용자의 목록 추천
*   유사도를 기반으로 추천
*  컨텐츠 기반 필터링  장점
  *   많은 수의 사용자를 대상으로 쉽게 확장 가능
  *   사용자가 관심을 갖지 않던 상품 추천 가능
*  컨텐츠 기반 필터링  단점
  *   입력 특성을 직접 설계해야 하기 때문에 많은 도메인 지식이 필요
  *   사용자의 기존 관심사항을 기반으로만 추천 가능









*   이전 백터의 내적을 통해 다른 사용자들과의 유사도 구하기
*   나와 가장 높은 유사도를 가진 사용자의 시청 목록을 추천

## 협업 필터링(Collaborative Filtering)

* 사용자와 항목의 유사성을 동시에 고려해 추천
* 기존에 내 관심사가 아닌 항목이라도 추천 가능
* 자동으로 임베딩 학습 가능
* 협업 필터링 장점
  * 자동으로 임베딩을 학습하기 때문에 도메인 지식이 필요 없음
  * 기존의 관심사가 아니더라도 추천 가능
* 협업 필터링 단점
  * 학습 과정에 나오지 않은 항목은 임베딩을 만들 수 없음
  * 추가 특성을 사용하기 어려움

* NMF를 사용한 협업 필터링

* SVD++를 사용한 협업 필터링

## 하이브리드(Hybrid)

* 컨텐츠 기반 필터링과 협업 필터링을 조합한 방식
* 많은 하이브리드 방식이 존재
* 실습에서는 협업 필터링으로 임베딩을 학습하고 컨텐츠 기반 필터링으로 유사도 기반 추천을 수행하는 추천 엔진 개발

* 사용자 기반 추천
* 나와 비슷한 취향을 가진 다른 사용자의 행동을 추천
* 사용자 특징 벡터의 유사도 사용

* 항목 기반 추천
* 내가 본 항목과 비슷한 항목을 추천
* 항목 특징 벡터의 유사도 사용

* 비음수 행렬 분해를 사용한 하이브리드 추천

* 사용자 기반 추천

* 항목 기반 추천

## 잠재 요인 협업 필터링(Latent Factor Collaborative Filtering)

* 사용자-아이템 평점 매트릭스 속에 숨어 있는 잠재 요인을 추출해 추천 예측을 할 수 있게 하는 기법
* 대규모 다차원 행렬을 SVD와 같은 차원 감소 기법으로 분해하는 과정에서 잠재 요인을 추출하는데, 이러한 기법을 행렬 분해(Matrix Factorization)라고 함

image.png

- 행렬 분해 기법을 이용해 사용자-잠재 요인 행렬과 아이템-잠재 요인 행렬의 전치 행렬(즉, 잠재 요인-아이템 행렬)로 분해된 데이터 세트를 다시 내적 곱으로 결합하면서 사용자가 예측하지 않은 아이템에 대한 평점을 도출하는 방식을 대략적으로 나타낸 것
- 행렬 분해에 의해 추출되는 '잠재 요인'이 정확히 어떤 것인지는 알 수 없지만, 가령 도서 평점 기반의 사용자-아이템 평점 행렬 데이터라면 도서가 가지는 장르별 특성 선호도로 가정할 수 있음
- 즉, 사용자-잠재 요인 행렬은 사용자의 도서 장르에 대한 선호도로, 아이템-잠재 요인 행렬은 도서의 장르별 특성 값으로 정의할 수 있음
