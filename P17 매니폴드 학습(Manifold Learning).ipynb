# 매니폴드 학습 (Manifold Learning)

* 고차원 데이터를 저차원으로 임베딩하는 비선형 차원 축소 기법의 일종
<br/>
* 데이터의 복잡한 구조와 패턴을 시각화하거나 데이터의 특성을 추출하는 데 사용됨
<br/>
* 고차원 데이터는 일반적으로 특성의 수가 많거나 복잡한 구조를 가지고 있어 데이터를 그대로 사용하면 분석이 어렵고 시각화도 어려움
* 이러한 고차원 데이터의 복잡한 구조를 적절하게 저차원 공간으로 임베딩하여 데이터를 시각화하거나 분석에 활용할 수 있도록 하는 것
* 로 비선형적인 데이터 구조를 다루는데 효과적이며, 데이터의 시각화, 특성 추출, 압축 등 다양한 응용 분야에서 활용
* 데이터의 복잡성과 차원의 저주에 따라 적절한 알고리즘과 매개변수 선택이 중요하며, 결과의 해석에 주의가 필요

### Locally Linear Embeddings (LLE)

* 최적의 비선형 임베딩을 찾기 위해 전체적으로 비교되는 일련의 로컬 주성분 분석
* 데이터 포인트 간의 지역적인 패턴을 보존하며 저차원 임베딩을 수행


**LLE 알고리즘의 주요단계**
1.   이웃 선택 : 각 데이터 포인트에 대해 근접한 이웃을 선택합니다. 이웃의 개수는 사용자가 지정
2.   근접 이웃 간의 선형 관계 확인 : 선택한 이웃들간의 선형관계를 확인하여 각 데이터 포인트를 근접 이웃들로 어떻게 잘 설명할 수 있는지를 파악하고, 이를 위해 선형 회귀를 사용하여 근접 이웃들을 가중치로 선형 조합하여 원래 데이터 포인트를 재구성
3.   임베딩: 재구성 오차를 최소화하는 저차원 임베딩을 찾아서 임베딩 공간에서 각 데이터 포인트는 근접 이웃들의 선형 조합으로 표현하고, 최종적으로 얻은 임베딩은 저차원에서의 데이터 포인트들 간의 구조를 보존하면서 고차원 데이터를 잘 표현


**LLE 알고리즘의 장점**
*   LLE은 데이터의 지역적인 패턴 구조를 잘 보존하며, 비선형 데이터의 특성을 잘 반영할 수 있습니다.
*   또한, 전역적인 데이터 구조를 고려하지 않으므로 매우 비선형적인 데이터에 적합한 결과를 제공할 수 있습니다.

**LLE 알고리즘의 단점**
*   그러나 LLE은 전역적인 구조를 완벽하게 보존하지는 않으며, 임베딩 결과가 초기 이웃 선택에 영향을 받을 수 있습니다
*   또한, 이웃 선택과 선형 회귀를 수행하는 과정에서 계산 비용이 증가할 수 있습니다.







### Isomap Embedding

Isometric Mapping을 통한 비선형 차원 감소
Isomap은 모든 지점 사이의 측지선 거리를 유지하는 저차원 임베딩을 찾음

* 데이터 포인트들 간의 거리 정보를 활용하여 저차원 공간에 매핑
* 데이터 포인트 간의 거리 정보를 보존하며 차원을 축소하기 때문에, 데이터 포인트들의 전역적인 구조를 고려
* 이를 위해 Isomap은 데이터 포인트들 사이의 최단 경로 거리를 계산합니다. 이 최단 경로 거리는 원본 데이터의 거리 정보를 기반으로 생성되며, 이를 통해 데이터 포인트들의 지역적인 패턴 구조를 무시하지 않고 고려

**Isomap Embedding 장점**
*   전역적인 데이터 구조 보존: Isomap은 데이터 포인트 간의 최단 경로 거리를 기반으로 임베딩을 수행하기 때문에, 데이터의 전역적인 구조를 고려, 이는 데이터의 복잡한 구조를 잘 보존할 수 있게 해줌
*   비선형 데이터 임베딩 : Isomap은 비선형적인 데이터에도 적용할 수 있는 임베딩 기법이며, 선형적인 임베딩 방법에 비해 더 유연하게 데이터의 구조를 반영할 수 있음
*   매개변수 설정의 상대적인 용이성 :  Isomap은 주요 매개변수인 최단 경로 거리 계산을 위한 이웃의 수를 설정해야 하고, 이 이웃의 수에 따라 결과가 달라질 수 있지만, 일반적으로 이웃의 수에 민감하지 않기 때문에 매개변수 설정이 다른 알고리즘에 비해 상대적으로 용이

**Isomap Embedding 단점**
*   계산 복잡성: Isomap은 최단 경로 거리 계산을 위해 그래프를 구성해야 하므로, 데이터의 크기가 큰 경우 계산 비용이 증가할 수 있고, 특히 데이터 포인트의 개수가 많거나 고차원 데이터인 경우에는 계산에 많은 시간과 자원이 소요될 수 있음
*   매개변수 선택의 어려움: 최단 경로 거리 계산을 위한 이웃의 수를 설정해야 하는데, 이 이웃의 수를 어떻게 선택해야 하는지가 어려울 수 있고, 이웃의 수가 적절하지 않게 설정되면 결과 임베딩에 왜곡이 발생할 수 있음
*   선형적 구조에는 한계: Isomap은 비선형 데이터에 더 적합하며, 선형적인 구조의 데이터에는 다른 선형 임베딩 알고리즘이 더 나은 결과를 제공할 수 있음








### Multidimensional Scaling (다차원 척도법)

다차원 척도법(Multi-dimensional Scaling, MDS)은 고차원의 데이터를 저차원으로 변환하는 기법으로 데이터 간의 유사성을 보존하면서 데이터를 저차원 공간에 임베딩하는 방법

**MDS의 작동 단계**
1.   데이터 유사성 측정: MDS는 주어진 데이터에서 측정하고자 하는 유사성 또는 거리 행렬을 생성 (예를 들어, 유클리드 거리, 상관관계, 유사도 등을 사용하여 데이터 간의 유사성을 측정할 수 있음)
2.   거리 행렬 변환: 생성된 유사성 또는 거리 행렬을 이용하여 거리 행렬을 만들고, 이 거리 행렬은 고차원 데이터의 각 데이터 포인트 간의 거리를 나타냄
3.   저차원 임베딩: 생성된 거리 행렬을 기반으로 MDS는 저차원 공간에 데이터를 임베딩하고, 이 임베딩은 고차원 데이터의 구조를 보존하려고 하며, 가능한 한 원래 데이터의 거리를 유지하려고 함
4.   좌표 계산: MDS는 임베딩된 데이터의 좌표를 계산하고, 일반적으로 이는 저차원 좌표로 표현되며, 각 좌표는 해당 데이터 포인트의 임베딩된 위치를 나타냄

**MDS 목표**
- 차원 데이터의 구조를 저차원으로 보존하는 것이고, 이를 통해 데이터를 시각화하거나 비교 분석을 수행할 수 있음
- MDS는 데이터 간의 관계를 시각적으로 이해하기 쉽게 만들어줌

**MDS 변형**
- 등간척도법(Metric MDS)과 비등간척도법(Non-metric MDS)이 있음
- 등간척도법은 거리 행렬을 정확하게 재구성하는 반면, 비등간척도법은 거리의 상대적인 비교만을 보존
- 등간척도법은 실제 거리 정보가 중요한 경우에 유용하며, 비등간척도법은 상대적인 유사성 정보를 고려할 때 유용

**MDS 장점**
- 데이터 시각화: MDS는 고차원 데이터를 저차원으로 임베딩하여 시각화할 수 있고, 이를 통해 데이터의 구조와 패턴을 시각적으로 이해할 수 있음
- 유사성 보존: MDS는 데이터 간의 유사성을 보존하려고 하며, 임베딩된 공간에서도 유사한 데이터 간의 거리를 유지하고, 이는 데이터 간의 상대적인 관계를 보존하고 비교 분석을 수행하는 데 도움이 됨
- 비선형 관계 탐색: MDS는 비선형적인 관계를 탐색할 수 있고, 데이터의 복잡한 구조를 잘 반영하며, 고차원 데이터의 비선형성을 저차원으로 투영할 수 있음

**MDS 단점**
- 계산 복잡성: MDS는 대규모 데이터셋에 대해 계산적으로 비용이 많이 들 수 있고, 특히, 거리 행렬의 계산과 저차원 임베딩을 위한 최적화 과정이 시간과 자원을 많이 소비
- 차원 선택: MDS에서는 저차원으로의 임베딩을 위해 목표 차원을 사전에 정해야하고, 적절한 차원을 선택하는 것은 중요한 문제이며, 잘못된 차원 선택은 데이터의 구조를 왜곡시킬 수 있음
- 이상치 민감성: MDS는 이상치에 민감할 수 있으며, 이상치는 거리 행렬을 왜곡시키고 임베딩 결과에 영향을 줄 수 있으므로, 이상치 처리에 주의해야 함








### Spectral Embedding

비선형 차원 축소를 위한 방법 중 하나로 데이터를 고차원의 유클리드 공간에서 낮은 차원의 임베딩 공간으로 매핑하는 기법

**Spectral Embedding의 작동 단계**
1.   데이터 유사도 행렬 생성: 먼저, 데이터 포인트 간의 유사도를 계산하여 유사도 행렬을 생성 (일반적으로는 거리 행렬을 사용하며, 주로 가우시안 유사도 함수를 이용하여 유사도를 계산)
2.   그래프 생성: 유사도 행렬을 기반으로 그래프를 생성하고, 보통 k-최근접 이웃(k-nearest neighbors) 또는 이웃 반경(neighborhood radius)을 설정하여 데이터 포인트 간의 연결을 정의
3.   그래프 라플라시안 계산: 생성된 그래프를 기반으로 그래프 라플라시안(Laplacian) 행렬을 계산하고, 그래프 라플라시안은 그래프의 구조와 연결성을 포착하는 중요한 정보를 담고 있음
4.   고유값 분해: 계산된 그래프 라플라시안 행렬을 고유값 분해하고, 이를 통해 고유값과 고유벡터를 얻을 수 있음
5.   임베딩 공간 선택: 고유값 분해 결과로부터 일부 고유값과 해당하는 고유벡터를 선택하고, 선택된 고유값과 고유벡터는 낮은 차원의 임베딩 공간을 형성
6.  임베딩 공간으로의 매핑: 선택된 고유벡터를 이용하여 데이터를 임베딩 공간으로 매핑하고, 이를 통해 원래 데이터의 비선형적인 구조를 잘 보존하면서 저차원으로 투영된 표현을 얻을 수 있음


**Spectral Embedding 장점**
1.   비선형 구조 보존: Spectral Embedding은 비선형적인 데이터 구조를 잘 보존할 수 있고, 이는 데이터의 복잡한 관계를 잘 표현할 수 있으며, 시각화나 비교 분석에 유용
2.   차원 축소 효과적: 낮은 차원의 임베딩 공간으로 데이터를 투영하는데 효과적이고, 이를 통해 고차원 데이터의 시각화와 해석이 용이
3.   유사도 정보 활용: Spectral Embedding은 데이터의 유사도 정보를 활용하여 임베딩을 수행하고, 이는 데이터 간의 관계를 고려하며, 이웃 데이터 간의 관계를 보존하는 임베딩을 생성할 수 있음

**Spectral Embedding 단점**

1.   계산 복잡성: Spectral Embedding은 그래프 생성과 그래프 라플라시안 계산 등의 계산적인 비용이 크고, 특히, 데이터셋이 크거나 밀도가 높은 경우에는 계산 시간과 메모리 사용량이 증가할 수 있음
2.   하이퍼파라미터 설정: Spectral Embedding에는 그래프 생성 및 라플라시안 계산과 관련된 하이퍼파라미터가 있고, 이를 적절하게 설정해야만 좋은 결과를 얻을 수 있으며, 하이퍼파라미터 설정에 대한 경험과 도메인 지식이 필요
3.   선형 분리 불가능: Spectral Embedding은 비선형 데이터에 대해 좋은 성능을 보이지만, 선형적으로 분리되는 데이터에는 제한적이고, 선형 구조에 대해서는 다른 선형 차원 축소 알고리즘이 더 효과적일 수 있음
4.   그래프 생성에 따른 결과 변동성: Spectral Embedding은 그래프 생성에 따라 결과가 달라질 수 있고, 그래프 생성 방법과 그래프 파라미터 설정에 따라 임베딩 결과가 달라질 수 있으므로, 주의가 필요








### T-distributed Stochastic Neighbor Embedding (t-SNE)

비선형 차원 축소 알고리즘으로, 고차원 데이터의 시각화에 많이 사용
데이터의 국부적인 구조를 보존하며, 저차원 임베딩을 생성하는데 초점을 둠

**t-SNE의 주요 아이디어**

- 유사한 데이터 포인트 간의 거리를 보존하려는 목표: 고차원 공간에서 유사한 데이터 포인트는 저차원 임베딩 공간에서도 가까이 위치하도록 함
- 차원 축소를 위한 확률 분포 모델링: t-SNE는 고차원 공간에서의 데이터 간 거리를 확률 분포로 모델링하고, 저차원 공간에서의 거리를 다른 확률 분포로 모델링

**t-SNE의 작동 방식**

1.   고차원 데이터 간의 유사도 계산: 먼저 고차원 데이터 포인트 간의 유사도를 계산 (일반적으로 유클리디안 거리를 사용하거나, 다른 유사도 척도를 선택할 수 있음)
2.   확률 분포 모델링: 계산된 유사도를 기반으로 고차원 공간에서의 유사도를 확률 분포로 모델링하고, 각 데이터 포인트를 중심으로 가우시안 분포를 생성하고, 이를 이웃 데이터 포인트와의 유사도에 기반하여 가중치를 부여
3.   저차원 임베딩 생성: 저차원 임베딩 공간에서의 유사도를 모델링하기 위해 쿨롱-레바넨젠 스탑그래드 알고리즘을 사용하고, 이 알고리즘은 고차원과 저차원 공간 간의 유사도를 최대한 일치시키면서 임베딩을 생성
4.   Gradient descent를 사용한 최적화: 저차원 임베딩을 생성하는 과정에서 Gradient descent를 사용하여 최적화하고, 이를 통해 임베딩의 안정성과 국부 구조의 보존을 개선할 수 있음
5.   시각화: 최적화된 저차원 임베딩을 시각화하여 데이터의 국부적 구조 및 군집을 살펴볼 수 있음

**t-SNE 장점**
1.   비선형 구조 시각화: t-SNE는 고차원 데이터의 비선형 구조를 잘 시각화할 수 있고, 군집이나 클러스터 간의 거리가 가까운 포인트들이 잘 분리되어 시각적으로 나타냄
2.   파라미터 설정의 유연성: t-SNE는 perplexity 값과 학습률 등의 파라미터를 조정하여 시각화 결과를 세밀하게 제어할 수 있음


**t-SNE 단점**
1.   시간 복잡도와 계산 부하: t-SNE는 고차원 데이터의 경우 계산이 매우 복잡하고 시간이 오래 걸릴 수 있고, 특히 큰 데이터셋의 경우에는 계산 부하가 크고 적절한 하이퍼파라미터 튜닝이 필요
2.   임베딩 안정성: t-SNE는 초기 임베딩 설정에 영향을 받을 수 있고, 같은 데이터에 대해 여러 번 실행하면 다른 결과를 얻을 수 있음
3.   거리 보존의 한계: t-SNE의 임베딩 결과에서 포인트 간의 거리는 상대적인 정보를 제공하며, 정확한 거리 척도로 사용하기에는 제한이 있음

Swiss Roll 데이터 셋 생성

 * 반면에 t-SNE는 데이터의 일반적인 구조를 보존할 수 있지만 원본 데이터의 연속적인 특성을 제대로 나타내지 못함 대신 포인트 섹션을 불필요하게 뭉치는 것처럼 보임

### Swiss-Hole

### Swiss-Hole


* LLE 및 t-SNE 임베딩을 계산하면 Swiss Roll과 유사한 결과를 얻음, LLE는 데이터를 매우 유능하게 펼치고 홀을 보존<br/>
* t-SNE는 포인트 섹션을 함께 묶는 것처럼 보이지만 원래 데이터의 일반적인 토폴로지를 보존

## t-SNE: 다양한 perplexity 값이 모형에 미치는 영향

- 서로 다른 perplexity 값에 대한 두 개의 동심원 및 S-커브 데이터 세트에 대한 t-SNE의 시각화
- perplexity 값이 증가할수록 도형이 더 선명해지는 경향이 관찰
- 클러스터의 크기, 거리 및 모양은 초기화, 난해도 값에 따라 달라질 수 있으며 항상 의미를 전달하는 것은 아님
- 복잡도가 높을 때 t-SNE는 두 개의 동심원으로 구성된 의미 있는 토폴로지를 찾지만 원의 크기와 거리는 원본과 약간 다름
- 두 개의 원 데이터 세트와는 달리, 더 큰 복잡도 값에 대해서도 S-커브 데이터 세트의 S-커브 토폴로지에서 시각적으로 모양이 달라짐
- 't-SNE를 효과적으로 사용하는 방법'(https://distill.pub/2016/misread-tsne/)

* 서로 다른 perplexity 값에 대한 두 개의 동심원 및 S-곡선 데이터 세트의 t-SNE 그림
* Perplexity 값이 증가함에 따라 더 선명한 모양을 향한 경향을 관찰
